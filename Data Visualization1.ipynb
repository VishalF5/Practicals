```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
totalDeath=[113055,37312,59176,7413,33694]
totalRecovery=[773480,32506,23659,166652,50473]
activeCases=[30000,56249,36425,21563,41365]
country=['USA','India','Brazil','Canada','England']
X=np.arange(len(totalDeath))

plt.bar(X,totalDeath,color='red',width=0.25)
plt.bar(X+0.25,totalRecovery,color='blue',width=0.25)
plt.bar(X+0.5,activeCases,color='green',width=0.25)
plt.legend(['Total Deaths','Total Recovery','Active Cases'])

plt.xticks([1+0.25 for i in range(5)],country)
plt.xlabel('Countries')
plt.ylabel('Cases')
plt.show()
```

```python
from pandas import DataFrame
from matplotlib import pyplot as plt
Data={
    'Tasks':[300,500,700],
    'Task Type':['Task Pending','Task Ongoing','Task Completed']
}
df=DataFrame(Data)
df.set_index('Task Type',inplace=True)
df.plot.pie(y='Tasks',autopct='%1.1f%%',startangle=90)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

labels=['Civil','Electical','Mechanical','Chemical']
sizes=[15,30,45,10]
explode=(0.1,0.1,0.1,0.4)
fig,ax=plt.subplots()
ax.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)
ax.axis('equal')
ax.set_title('Engineering Disciplines')
```

# Subplots

```python
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.bar(range(1,6),np.random.randint(1,20,5))
plt.title('2,2,1')
plt.subplot(2,2,2)
plt.bar(range(1,6),np.random.randint(1,20,5))
plt.title('2,2,2')
plt.subplot(2,2,3)
plt.scatter(range(1,6),np.random.randint(1,20,5),s=100,color='r')
plt.title('2,2,3')
plt.subplot(2,2,4)
plt.plot(range(1,6),np.random.randint(1,20,5),marker='o',color='g',linestyle='--')
plt.title('2,2,4')
```
